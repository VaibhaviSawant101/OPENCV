```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("./RetinaDataset/Severe.png")
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("GRAY", gray)
cv.waitKey(0)
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow("HSV", hsv)
cv.waitKey(0)
```

```python
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow("LAB", lab)
cv.waitKey(0)
```

```python
plt.imshow(img)
plt.show()
```

```python
cv.imshow("img", img)
cv.waitKey(0)
```

```python
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("rgb", rgb)
cv.waitKey(0)
```

```python
bgr = cv.cvtColor(hsv, cv.COLOR_HSV2BGR)
cv.imshow("bgr", bgr)
cv.waitKey(0)
```

```python
bgr = cv.cvtColor(lab, cv.COLOR_LAB2BGR)
cv.imshow("bgr", bgr)
cv.waitKey(0)
```
